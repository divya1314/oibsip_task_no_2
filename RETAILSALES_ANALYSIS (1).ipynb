# **Nutrition Facts for McDonald's Menu**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Data Loading and Cleaning
df = pd.read_csv('/content/menu.csv')
print("First few rows of the dataset:")
print(df.head())

print("\nColumn names in the dataset:")
print(df.columns)

print("\nMissing values in each column:")
print(df.isnull().sum())
df = df.dropna()

# Step 2: Descriptive Statistics

mean_calories = df['Calories'].mean()
median_calories = df['Calories'].median()
mode_calories = df['Calories'].mode()[0]
std_calories = df['Calories'].std()

print(f"\nMean Calories: {mean_calories}")
print(f"Median Calories: {median_calories}")
print(f"Mode Calories: {mode_calories}")
print(f"Standard Deviation of Calories: {std_calories}")


mean_fat = df['Total Fat'].mean()
mean_carbs = df['Carbohydrates'].mean()
mean_protein = df['Protein'].mean()

print(f"\nMean Total Fat: {mean_fat}")
print(f"Mean Carbohydrates: {mean_carbs}")
print(f"Mean Protein: {mean_protein}")

# Step 3: Nutritional Analysis by Category

category_nutrition = df.groupby('Category').agg({
    'Calories': 'mean',
    'Total Fat': 'mean',
    'Carbohydrates': 'mean',
    'Protein': 'mean'
}).reset_index()

print("\nAverage nutritional values by category:")
print(category_nutrition)

# Step 4: Best and Worst Foods for Calories and Fat

# Food with highest and lowest calories
max_calories_food = df.loc[df['Calories'].idxmax()]
min_calories_food = df.loc[df['Calories'].idxmin()]

print("\nFood item with the highest calories:")
print(max_calories_food[['Item', 'Category', 'Calories', 'Total Fat', 'Carbohydrates', 'Protein']])

print("\nFood item with the lowest calories:")
print(min_calories_food[['Item', 'Category', 'Calories', 'Total Fat', 'Carbohydrates', 'Protein']])

# Food with highest and lowest total fat
max_fat_food = df.loc[df['Total Fat'].idxmax()]
min_fat_food = df.loc[df['Total Fat'].idxmin()]

print("\nFood item with the highest total fat:")
print(max_fat_food[['Item', 'Category', 'Calories', 'Total Fat', 'Carbohydrates', 'Protein']])

print("\nFood item with the lowest total fat:")
print(min_fat_food[['Item', 'Category', 'Calories', 'Total Fat', 'Carbohydrates', 'Protein']])

# Step 5: Visualization

plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Calories', data=category_nutrition)
plt.title('Average Calories by Menu Category')
plt.xlabel('Category')
plt.ylabel('Average Calories')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Calories', data=df)
plt.title('Calories Distribution by Menu Category')
plt.xlabel('Category')
plt.ylabel('Calories')
plt.xticks(rotation=45)
plt.show()

corr_matrix = df[['Calories', 'Total Fat', 'Carbohydrates', 'Protein']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Nutritional Values')
plt.show()

# Step 6: Recommendations
recommendations = """
1. Highlight low-calorie options like Beverages and Salads.
2. Introduce balanced items with reduced calories and fat in high-demand categories like Chicken & Fish and Breakfast.
3. Promote high-protein items in Beef & Pork and Chicken & Fish categories.
4. Reduce total fat in popular items to align with health trends.
5. Develop healthier versions of Smoothies & Shakes with less sugar and more protein.
6. Ensure transparency of nutritional information and educate customers on healthier choices.
7. Introduce seasonal and limited-time healthy promotions.
"""

print("\nRecommendations based on EDA:")
print(recommendations)
```

# **RETAIL ANALYSIS WITH WALMART SALES DATA**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Data Loading and Cleaning

df = pd.read_csv('/content/WALMART_SALES_DATA.csv')

print("First few rows of the dataset:")
print(df.head())

print("\nColumn names in the dataset:")
print(df.columns)


print("\nMissing values in each column:")
print(df.isnull().sum())


df = df.dropna()

df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")  

# Step 2: Descriptive Statistics

mean_sales = df['Weekly_Sales'].mean()
median_sales = df['Weekly_Sales'].median()
mode_sales = df['Weekly_Sales'].mode()[0]
std_sales = df['Weekly_Sales'].std()

print(f"\nMean Weekly Sales: {mean_sales}")
print(f"Median Weekly Sales: {median_sales}")
print(f"Mode Weekly Sales: {mode_sales}")
print(f"Standard Deviation of Weekly Sales: {std_sales}")


mean_temperature = df['Temperature'].mean()
mean_fuel_price = df['Fuel_Price'].mean()
mean_cpi = df['CPI'].mean()
mean_unemployment = df['Unemployment'].mean()

print(f"\nMean Temperature: {mean_temperature}")
print(f"Mean Fuel Price: {mean_fuel_price}")
print(f"Mean CPI: {mean_cpi}")
print(f"Mean Unemployment: {mean_unemployment}")

# Step 3: Analysis by Store and Region

store_sales = df.groupby('Store').agg({
    'Weekly_Sales': 'mean'
}).reset_index()

print("\nAverage weekly sales by store:")
print(store_sales)

# Step 4: Impact of Holidays and Markdowns

holiday_sales = df.groupby('Holiday_Flag').agg({
    'Weekly_Sales': 'mean'
}).reset_index()

print("\nAverage weekly sales during holidays vs. non-holidays:")
print(holiday_sales)

# Step 5: Best and Worst Performing Stores in Terms of Sales


max_sales_store = df.loc[df['Weekly_Sales'].idxmax()]
min_sales_store = df.loc[df['Weekly_Sales'].idxmin()]

print("\nStore with the highest weekly sales:")
print(max_sales_store[['Store', 'Date', 'Weekly_Sales', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']])

print("\nStore with the lowest weekly sales:")
print(min_sales_store[['Store', 'Date', 'Weekly_Sales', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']])

# Step 6: Visualization


plt.figure(figsize=(12, 6))
sns.barplot(x='Store', y='Weekly_Sales', data=store_sales)
plt.title('Average Weekly Sales by Store')
plt.xlabel('Store')
plt.ylabel('Average Weekly Sales')
plt.xticks(rotation=90)
plt.show()


plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Weekly_Sales', data=df)
plt.title('Weekly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(12, 6))
sns.boxplot(x='Holiday_Flag', y='Weekly_Sales', data=df)
plt.title('Weekly Sales Distribution During Holidays vs. Non-Holidays')
plt.xlabel('Holiday Flag')
plt.ylabel('Weekly Sales')
plt.xticks(rotation=45)
plt.show()

corr_matrix = df[['Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Values')
plt.show()

# Step 7: Recommendations

recommendations = """
1. Monitor and stock up inventory for holidays to prevent stockouts.
2. Focus on stores with high sales for targeted promotions.
3. Adjust inventory based on economic indicators like CPI and unemployment.
4. Use historical sales data to forecast and adjust for future demands.
5.Holiday Flag = 0: Represents non-holidays. The average weekly sales during non-holidays is 1,041,255.57.
Holiday Flag = 1: Represents holidays. The average weekly sales during holidays is 1,122,887.96.
Therefore, on average, the sales during holidays are higher compared to non-holidays.
Increase marketing efforts around holidays to capitalize on higher sales.
"""

print("\nRecommendations based on EDA:")
print(recommendations)
```
